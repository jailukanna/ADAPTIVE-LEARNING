```python
import sys, os
sys.path.append(os.path.abspath("src"))
import time

from src.puzzle_generator import generate_puzzle
from src.tracker import PerformanceTracker
from src.adaptive_engine import adjust_difficulty

tracker = PerformanceTracker()
name = input("Enter your name: ")
level = input("Choose initial difficulty (Easy/Medium/Hard): ").capitalize()

for i in range(5):  # simulate a session
    question, answer = generate_puzzle(level)
    print(f"\nPuzzle {i+1}: {question}")
    start = time.time()
    try:
        user_ans = float(input("Your answer: "))
    except ValueError:
        user_ans = None
    end = time.time()

    correct = abs(user_ans - answer) < 0.01 if user_ans is not None else False
    tracker.log(correct, end - start)
    print("✅ Correct!" if correct else f"❌ Wrong! Answer = {answer}")

summary = tracker.summary()
print("\nSession Summary:")
print(summary)

new_level = adjust_difficulty(level, summary["accuracy"])
print(f"Next recommended level: {new_level}")
```
